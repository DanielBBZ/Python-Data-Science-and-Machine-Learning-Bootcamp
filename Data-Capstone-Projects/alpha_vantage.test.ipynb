```python
from alpha_vantage.timeseries import TimeSeries
import matplotlib.pyplot as plt
%matplotlib inline

ts = TimeSeries(key='3WA8H4TVHIDTGIMP', output_format='pandas')
```

```python
#Time Series
#Using pandas support we can plot the intra-minute value for 'MSFT' stock quite easily:
data, meta_data = ts.get_intraday(symbol='MSFT',interval='1min',outputsize='full')
data['4. close'].plot()
plt.title('Intraday Times Series for the MSFT stock (1 min)')
```

```python
#Technical indicators
#The same way we can get pandas to plot technical indicators like Bolliger BandsÂ®
from alpha_vantage.techindicators import TechIndicators
import matplotlib.pyplot as plt

ti = TechIndicators(key='3WA8H4TVHIDTGIMP', output_format='pandas')
data, meta_data = ti.get_bbands(symbol='MSFT', interval='60min', time_period=60)
data.plot()
plt.title('BBbands indicator for  MSFT stock (60 min)')
```

```python
#Sector Performance
#We can also plot sector performance just as easy:

from alpha_vantage.sectorperformance import SectorPerformances
import matplotlib.pyplot as plt

sp = SectorPerformances(key='3WA8H4TVHIDTGIMP', output_format='pandas')
data, meta_data = sp.get_sector()
data['Rank A: Real-Time Performance'].plot(kind='bar')
plt.title('Real Time Performance (%) per Sector')
plt.tight_layout()
plt.grid()
```

```python
#Crypto currencies.
#We can also plot crypto currencies prices like BTC:

from alpha_vantage.cryptocurrencies import CryptoCurrencies
import matplotlib.pyplot as plt

cc = CryptoCurrencies(key='3WA8H4TVHIDTGIMP', output_format='pandas')
data, meta_data = cc.get_digital_currency_intraday(symbol='BTC', market='CNY')
data['1b. price (USD)'].plot()
plt.tight_layout()
plt.title('Intraday value for bitcoin (BTC)')
plt.grid()
```

```python
#Foreign Exchange (FX)
#The foreign exchange is just metadata, thus only available as json format (using the 'csv' or 'pandas' format will raise an Error)

from alpha_vantage.foreignexchange import ForeignExchange
from pprint import pprint
cc = ForeignExchange(key='3WA8H4TVHIDTGIMP')
# There is no metadata in this call
data, _ = cc.get_currency_exchange_rate(from_currency='BTC',to_currency='USD')
pprint(data)
```
